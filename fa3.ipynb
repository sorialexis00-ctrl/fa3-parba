```python
# Program No. 2 (fa3prog2)

score = float(input("Enter student's exam score: "))

if score >= 75:
    print("Passed")
else:
    print("Failed")
```

```python
# Program No. 3 (fa3prog3)

name = input("Enter student's name: ")
score = float(input("Enter student's score: ")

if 90 <= score <= 100:
    grade = "A"
elif 80 <= score <= 89:
    grade = "B"
elif 70 <= score <= 79:
    grade = "C"
else:
    grade = "F"

print("\n--- Grading Result ---")
print(f"Name  : {name}")
print(f"Score : {score}")
print(f"Grade : {grade}")
```
